### Diferencias de horario de Arribo (retrasos) ‚è∞

  Se desea analizar si hubo retrasos o demoras en la llegada de los vuelos. Para ello se comparan el horario estimado de arribo (STA = Scheduled Time Arrival) con el horario real de arribo (ATA = Actual Time Arrival) y se determina el tiempo de demora en minutos.

```python
df_Vuelos['Convert'] = (df_Vuelos['ATA'])-(df_Vuelos['STA']) # Devuelve un timedelta
df_Vuelos['Demora_Min'] = df_Vuelos['Convert'].dt.seconds.div(60)+df_Vuelos['Convert'].dt.days.multiply(1440) # Convierto el timedelta en minutos
df_Vuelos['Delayed'] = (df_Vuelos['Convert'].dt.seconds.div(60)+df_Vuelos['Convert'].dt.days.multiply(1440))>0 # Convierto el timedelta en booleano
df_Vuelos.Delayed = pd.to_numeric(df_Vuelos.Delayed).astype(int) # Convierto el booleano en un nro entero para poder sumarlo
df_Vuelos[['Delayed','ATA','STA','Convert','Demora_Min']] # Verifico las columnas creadas
```

```python
# Vamos a remover los outliers
def removeOutliers(df,col):

    Q3 = np.quantile(df[col], 0.75)
    print('quartil 3 es ', Q3)

    Q1 = np.quantile(df[col], 0.25)
    print('quartil 1 es ',Q1)

    IQR = Q3 - Q1
    print('rango intercuartilico es ', IQR)

    lower_range = Q1 - 1.5 * IQR
    print('extremo inferior es ', lower_range)

    upper_range = Q3 + 1.5 * IQR
    print('extremo superior es ', upper_range)

    df = df.loc[(df[col] > lower_range) & (df[col] < upper_range),:]

    return df


Arg_Res1=removeOutliers(Arg_Res,'Demora_Min')
```

## **MVP "Minimum Viable Product" (o "Producto M√≠nimo Viable")**  



Se emplear√° este concepto (derivado de la metodolog√≠a √°gil) para describir la versi√≥n m√°s simple/b√°sica del proyecto que contiene las caracter√≠sticas esenciales necesarias para cumplir con su prop√≥sito principal.

Esta versi√≥n preliminar y simplificada del modelo de aprendizaje autom√°tico, tiene por objetivo  obtener resultados iniciales r√°pidamente y validar si la idea o enfoque propuesto es viable antes de invertir recursos significativos en su desarrollo completo.

Se entrega:


*   Algoritmo Simplificado: c√≥digo del modelo de aprendizaje autom√°tico o algoritmo de an√°lisis de datos b√°sico, para proporcionar resultados iniciales.

*   Funcionalidades Clave: se incluyen solo las funcionalidades esenciales necesarias para demostrar la viabilidad de la soluci√≥n.

*   Validaci√≥n R√°pida: se realizan pruebas y validaciones iniciales para evaluar si el enfoque tiene el potencial de resolver el problema.

Una vez que se demuestre que el MVP es viable y se generen resultados prometedores, se puede iterar y mejorar gradualmente, agregando caracter√≠sticas adicionales, ajustando el modelo y expandiendo la soluci√≥n. Este MVP es una forma eficaz de reducir el riesgo y garantizar que los recursos se utilicen de manera eficiente en el desarrollo de una soluci√≥n m√°s completa y efectiva.
